```python
import pandas as pd

from sklearn import base
from sklearn import model_selection

random_state = 42
```

```python
X = pd.read_csv('data/train-cleaned.csv')
y = X.pop('y')
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, random_state=random_state)
```

```python
class ColumnSelector(base.BaseEstimator, base.TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        assert isinstance(X, pd.DataFrame)

        try:
            return X[self.columns]
        except KeyError:
            cols_error = list(set(self.columns) - set(X.columns))
            raise KeyError("The DataFrame does not include the columns: %s" % cols_error)
```

```python
class TypeSelector(base.BaseEstimator, base.TransformerMixin):
    def __init__(self, dtype):
        self.dtype = dtype

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        assert isinstance(X, pd.DataFrame)
        return X.select_dtypes(include=[self.dtype])
```

```python
cs = ColumnSelector(columns=X)
cs.fit_transform(X).head()
```
